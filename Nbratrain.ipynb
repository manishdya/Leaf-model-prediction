```python
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import json
import sys
import pickle
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
class NAVIEBAYES:
  try:
    def __init__(self):
      self.df=sns.load_dataset('iris')
      self.df['species']=self.df['species'].map({'setosa':0,'versicolor':1,'virginica':2}).astype(int)
      self.X=self.df.iloc[:,:-1]
      self.y=self.df.iloc[:,-1]
      self.X_train,self.X_test,self.y_train,self.y_test=train_test_split(self.X,self.y,test_size=0.2,random_state=45)
      self.knn_obj1=GaussianNB()
      self.knn_obj1.fit(self.X_train,self.y_train)


    def Testperformance(self):
      self.cm1=confusion_matrix(self.y_train,self.knn_obj1.predict(self.X_train))
      self.cf1=classification_report(self.y_train,self.knn_obj1.predict(self.X_train))
      self.acs1=accuracy_score(self.y_train,self.knn_obj1.predict(self.X_train))

      print(f'confusion matrix: {self.cm1}')
      print(f'classification report: {self.cf1}')
      print(f'accuracy score: {self.acs1}')






  except Exception :
    exc_type,exc_msg,exc_line=sys.info()
    print(exc_type,exc_msg,exc_line)


if __name__=='__main__':

  obj1=NAVIEBAYES()
  obj1.Testperformance()
```
